```python
from sklearn import datasets
```

```python
db=datasets.load_diabetes()
```

```python
db
```

```python
db.keys()
db.data
db.target
```

```python
import pandas as pd
```

```python
file =pd.DataFrame(data=db.data,columns=db.feature_names)
file
```

```python
file.corr()
```

```python
file_inp=file.ix[:,:-1]
file_op=file.ix[:,-1]
```

```python
file.shape
```

```python
file['DiseaseProgression']=db.target
```

```python
import statsmodels.formula.api as ms
```

```python
md=ms.ols(formula="DiseaseProgression~sex+bmi+bp+s1+s2+s3+s4",data=file)
reg=md.fit()
reg.summary()
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(file_inp,file_op).score(file_inp,file_op)
```

```python
file["My_prediction"]=lr.predict(file_inp)
```
