```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('HRemp.csv')
data
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
data.describe()
```

```python
corr=data.corr()
corr=(corr)
plt.figure(figsize=(10, 10))
sns.heatmap(corr,
           xticklabels=corr.columns.values,
           yticklabels=corr.columns.values,cmap='Reds')
```

```python
Attrition_Rate=data.Attrition.value_counts()/len(data)
Attrition_Rate
```

```python
sns.barplot(x='Attrition',y='MonthlyIncome',hue='Gender',data=data,color='red').set_title('Employee Income Gender Distribution')
plt.figure(figsize=(10, 10))
plt.show()
```

```python
sns.boxplot(data['Gender'], data['MonthlyIncome'])
plt.title('MonthlyIncome vs Gender Box Plot', fontsize=20)      
plt.xlabel('MonthlyIncome', fontsize=16,color='black')
plt.ylabel('Gender', fontsize=16)
plt.show()
```

```python
sns.countplot(data.JobLevel)
plt.title('JobLevel Count Plot', fontsize=20)      
plt.xlabel('JobLevel', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.show()
```
